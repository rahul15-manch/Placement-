```python
""" 
1
22
333
4444
55555"""


n=6
for i in range(n):
    val=i
    for j in range(i):
        print(val,end="")
    print()
```

```python
""" 
1
10
101
1010
10101
"""
n=6
for i in range(n):
    for j in range(i):
        if j%2==0:
            print("1",end="")
        else:
            print("0",end="")
    print()
```

```python
"""
        *
       * *    
      * * *  
     * * * * 
    * * * * *
     
     """
n=6
for i in range(1,n):
    for space in range(n-i):
        print(" ",end="")
    for star in range(i):
        print("*",end=" ")
    print()
```

```python
n=6
for i in range(n-1,-1,-1):
    for space in range(n-i):
        print(" ",end="")
    for star in range(i):
        print("*",end=" ")
    print()
        
```
