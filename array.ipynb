```python
nums = [1,9,6,3,2,8]
# O/p=9,8

first = second = float('-inf')

for num in nums:
    if num > first:
        second = first
        first = num
    elif num > second:
        second = num

print(first, second)
```

```python
nums1 = [1,2,3,4]
nums2 = [3,4,5,6]

i = j = 0
res = []

while i < len(nums1) and j < len(nums2):

    if nums1[i] < nums2[j]:
        res.append(nums1[i])
        i += 1

    elif nums1[i] > nums2[j]:
        res.append(nums2[j])
        j += 1

    else:  
        i += 1
        j += 1

while i < len(nums1):
    res.append(nums1[i])
    i += 1

while j < len(nums2):
    res.append(nums2[j])
    j += 1

print(res)
```

```python
nums = [1,2,3,4,5,6,7]
target = 12

left = 0
curr_sum = 0

for right in range(len(nums)):
    curr_sum += nums[right]
    
    while curr_sum > target:
        curr_sum -= nums[left]
        left += 1
        
    if curr_sum == target:
        print(nums[left:right+1])
        break

        
```
