```python
# armstrong Number 
num=int(input("Enter a number"))
temp=num
sum=0
for i in range(len(str(num))):
    digit=temp%10
    sum+=digit**len(str(num))
    temp=temp//10

print(sum==num)
    
```

```python
# palindrome number 
num=121
temp=num
rev=0
for i in range(len(str(num))):
    digit=temp%10
    rev+=digit*(10**i)
    temp=temp//10
print(rev==num)
```

```python
# Neon Number
num=9
square=num**2
sum=0
for i in range(len(str(square))):
    digit=square%10
    sum+=digit
    square=square//10

print(sum==num)

```
